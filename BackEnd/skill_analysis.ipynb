```python
import spacy

def combine_skills(tokens):
    combined_skills = []
    current_skill = []

    for token in tokens:
        # Check if the token is part of a multi-word skill
        if token.pos_ == 'NOUN' or token.pos_ == 'ADJ':
            current_skill.append(token.text)
        else:
            # If the current skill is not empty, combine the tokens into a single skill
            if current_skill:
                combined_skills.append(' '.join(current_skill))
                current_skill = []
    
    # If there's any remaining skill at the end, add it to the combined list
    if current_skill:
        combined_skills.append(' '.join(current_skill))
    
    return combined_skills

# Example usage
nlp = spacy.load("en_core_web_sm")

# Tokenize the text
skill_list = ['demand', 'forecasting', 'inventory', 'management', 'data', 'analysis', 'sales', 'operations', 'planning', 'sop', 'erp', 'software', 'proficiency', 'construction', 'management']

tokens = nlp(skill_list)

# Combine tokens into multi-word skills
skills = combine_skills(tokens)

# Print the combined skills
print(skills)
```
